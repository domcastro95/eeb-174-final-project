```python
def makeEarlyIntervalList(filename): # defines and opens canis data file 
    EarlyIntervalList = [] #defines as a list 
    earlyinterval_object = open(filename,"r") #opens file 
    earlyinterval_list = earlyinterval_object.readlines()[19:] #reads file and save as list
    for line in earlyinterval_list:
        earlyinterval = line.split(",")[12]
        # stick intervals into EarlyIntervalList
        EarlyIntervalList.append(earlyinterval) #file time zone into late interval list 
    return EarlyIntervalList # run function 
```

```python
makeEarlyIntervalList("formatted-Canis_Occurances.csv")
```

```python
def makeMaxMinList(filename): # defines and opens canis data file 
    MaxMinList = [] #defines as a list 
    maxmin_object = open(filename,"r") #opens file 
    maxmin_list = maxmin_object.readlines()[19:] #reads file and save as list
    for line in maxmin_list:
        maxmin = line.split(",")[14:16]
        # stick periods into 
        MaxMinList.append(maxmin) #file time zone into late interval list 
    return MaxMinList # run function 
```

```python
# load in the file

in_file = open("/home/eeb177-student/Desktop/eeb-177/eeb-174-final-project/formatted-Canis_Occurances4.txt")
# read in the first line
# The first line is a header that we don't want in dictionaries
#in_file.readline()
# set up empty dictionaries
max_ma = {}
min_ma = {}

# iterate through every line in the file
for line in in_file:
    #print(line)
    
    # It's a comma delimited file, so we can split the contents up into 
    # the component rows
    items = line.split(",")
    
    # Save some important pieces of information as variables # change these 
    canis_occurance_no = items[0]
    max_ma_number = float(items[14])
    min_ma_number = float(items[15])
    
    # Use the variables to populate our three empty lists
    max_ma[canis_occurance_no] = max_ma_number
    min_ma[canis_occurance_no] = min_ma_number
```

```python
# Import in the pyplot function from matplotlib, and 
# make it available to us with the name "plt"
# NOTE: assumes you have installed matplotlib onto your virtual machine
# See Issue #17 on the course github repo for instructions on installation
import matplotlib.pyplot as plt


# Most plotting happens on lists, not dictionaries-
# so, let's convert our 1994 DBH meaasurements into a list
max_ma_number = list(max_ma.values())
min_ma_number = list(min_ma.values())

# We're ready to make a histogram!

plt.hist(max_ma_number)
plt.hist(min_ma_number)
# an X-axis label
plt.xlabel('Max age of Fossil (Ma)')
# a y-axis label
plt.ylabel('Frequency')
# A main title for the graph
plt.title("Number of fossils found in different years")
# Let's look!
plt.show()

```
